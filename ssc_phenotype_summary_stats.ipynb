### Replace "not calculated" entries with "NaN"
This way I can then convert the object types to float types and deal with all the NaNs for all columns at the same time.

"Not calculated for Module 4" is found in the ados_css and ados_social_affect columns

```python
ssc_phenos_df = ssc_phenos_df.replace('Not calculated for Module 4',np.NaN)
```

```python
ssc_phenos_df.head(7)
```

### Convert columns to to "float" types

```python
ssc_phenos_df["ados_css"] = ssc_phenos_df['ados_css'].astype('float')
ssc_phenos_df["ados_social_affect"] = ssc_phenos_df['ados_social_affect'].astype('float')
ssc_phenos_df["age_at_ados"] = ssc_phenos_df['age_at_ados'].astype('float')
```

```python
ssc_phenos_df.info()
```

### Replace NaNs with the mean value for each feature

```python
ssc_phenos_df.describe()
```

#### Replace NaN in a given column with its respective mean value, using a dictionary

```python
means = {'ados_css': ssc_phenos_df.ados_css.mean(),
         'ssc_diagnosis_verbal_iq': ssc_phenos_df.ssc_diagnosis_verbal_iq.mean(),
         'ssc_diagnosis_nonverbal_iq': ssc_phenos_df.ssc_diagnosis_nonverbal_iq.mean(),
         'ssc_diagnosis_full_scale_iq': ssc_phenos_df.ssc_diagnosis_full_scale_iq.mean(),
        'adi_r_b_comm_verbal_total': ssc_phenos_df.adi_r_b_comm_verbal_total.mean(),
         'adi_r_comm_b_non_verbal_total': ssc_phenos_df.adi_r_comm_b_non_verbal_total.mean(),
        'ados_restricted_repetitive':ssc_phenos_df.ados_restricted_repetitive.mean(),
        'adi_r_rrb_c_total':ssc_phenos_df.adi_r_rrb_c_total.mean(),
        'adi_r_soc_a_total':ssc_phenos_df.adi_r_soc_a_total.mean(),
        'ados_social_affect':ssc_phenos_df.ados_social_affect.mean(),
        'vineland_ii_composite_standard_score':ssc_phenos_df.vineland_ii_composite_standard_score.mean(),
        'rbs_r_overall_score':ssc_phenos_df.rbs_r_overall_score.mean(),
        'ados_communication_social':ssc_phenos_df.ados_communication_social.mean()}
```

```python
ssc_phenos_df = ssc_phenos_df.fillna(value=means)
ssc_phenos_df.info()
```
